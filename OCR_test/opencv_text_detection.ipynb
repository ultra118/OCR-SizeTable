```python
import sys
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

# gray 이미지로바꾸고  blur 먹이고 Threshhlod로 이진화

```python
img = cv2.imread('./image/crop_table2.PNG')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (9,9), 0)
thresh = cv2.adaptiveThreshold(blur, 255, 1, 1, 11, 2)
```

# 윤곽 추출하기
- 참조 : https://datascienceschool.net/view-notebook/f9f8983941254a34bf0fee42c66c5539/

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# contours : 컨투어 포인트들의 배열이 리스트로 되어있음
```

```python
print(f'contours len : {len(contours)}')
hierarchy
```

```python
contours[0].shape
```

```python
for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    if h > 50:
        continue
    red = (0, 0, 255)
    cv2.rectangle(img, (x,y), (x+w, y+h), red, 2)
```

```python
plt.imshow(img)
```
